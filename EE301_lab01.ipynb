```
from matplotlib import pyplot as plt

x = [500, 1000, 1500, 1700]

y = [0.8, 1.9, 4.2, 'inf']


fig, ax = plt.subplots()

ax.plot(x,y)

ax.annotate('Ruptura', xy=(1700, 'inf'), xytext=(1900, 'inf'),
            arrowprops=dict(facecolor='black', shrink=0.05),
           )

plt.title('Medidas VxI em materiais dielétricos [Prespon 0,19mm]')
plt.ylabel(u'Corrente [\u03bcA]')
plt.xlabel('Tensão [V]')
plt.show()
```

```
x = [500, 1000, 1500, 2000, 2500, 3000, 3100]
y = [1.2, 2.4, 3.9, 5.8, 8.5, 11.8, 'inf']

fig, ax = plt.subplots()

ax.plot(x,y)

ax.annotate('Ruptura', xy=(3100, 'inf'), xytext=(3500, 'inf'),
            arrowprops=dict(facecolor='black', shrink=0.05),
           )


plt.plot(x,y)
plt.title('Medidas VxI em materiais dielétricos [Filtro Comercial 0,26 mm]')
plt.ylabel(u'Corrente [\u03bcA]')
plt.xlabel('Tensão [V]')
plt.show()
```
