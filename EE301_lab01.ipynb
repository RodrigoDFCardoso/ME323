```
from matplotlib import pyplot as plt

x = [500, 1000, 1500, 1700]

y = [0.8, 1.9, 4.2, 'inf']


fig, ax = plt.subplots()

plt.rcParams["font.family"] = "Times New Roman"

ax.plot(x,y)

ax.annotate('Ruptura', xy=(1700, 'inf'), xytext=(1900, 'inf'),
            arrowprops=dict(facecolor='black', shrink=0.05),
           )

plt.title('Medidas VxI em materiais dielétricos [Prespon 0,19mm]')
plt.ylabel(u'Corrente [\u03bcA]')
plt.xlabel('Tensão [V]')
plt.show()
```
