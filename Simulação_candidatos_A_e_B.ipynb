```
from matplotlib import pyplot as plt
import random

plt.rcParams['figure.figsize'] = (20,10)
eleitores = []

a = 0
b = 0

Ca = 0
Cb = 0

lis_loop = []
lis_resultado = []
lis_resA = []
lis_resB = []

n = 8000

for i in range(10000):
	x = random.randrange(2)

	if (x == 0) and (Ca <= 3000):
		eleitores.append("A")
		Ca += 1

	else:
		if Cb <= 7000:
			eleitores.append("B")
			Cb += 1
      
histo_a = {}
histo_b = {}

def ger_prob(n, a , b, yg):

    for loop in range(2000):
      for iii in range(n):
        el = random.choice(eleitores)

        if el == "A":    
          a += 1
        elif el == "B":
          b += 1

      resultado_A = round(a/n, 3)
      resultado_B = round(b/n, 3)

      lis_resA.append(resultado_A)
      lis_resB.append(resultado_B)
      lis_loop.append(loop)

      if resultado_A not in histo_a:
        histo_a[resultado_A] = 1

      elif resultado_A in histo_a:
        histo_a[resultado_A] += 1

      if resultado_B not in histo_b:
        histo_b[resultado_B] = 1

      elif resultado_B in histo_b:
        histo_b[resultado_B] += 1

      a, b = 0, 0

    y_label_a = []
    x_label_a = []

    y_label_b = []
    x_label_b = []

    for res_a in histo_a:
      x_label_a.append(float(res_a))
      y_label_a.append(histo_a[res_a])

    for res_b in histo_b:
      x_label_b.append(float(res_b))
      y_label_b.append(histo_b[res_b])
      
    soma_a = soma_b = mult_b = mult_a = 0

    if len(y_label_a) == len(x_label_a):
      for s in range(len(y_label_a)):
        soma_a += y_label_a[s]
        mult_a += x_label_a[s] * y_label_a[s]

      result_a = round(mult_a / soma_a, 4)

    result_a_graf = r'$\overline{x} = $' + str(result_a)
    
    for ss in x_label_a:
      soma_desv_a =+ (ss - result_a) ** 2

    dp_a = round((soma_desv_a / len(x_label_a)) ** (1/2), 6)
    
    dp_a_graf = r'$ \sigma = $' + str(dp_a)

    if len(y_label_b) == len(x_label_b):
      for s in range(len(y_label_b)):
        soma_b += y_label_b[s]
        mult_b += x_label_b[s] * y_label_b[s]

      result_b = round(mult_b / soma_b, 4)

    result_b_graf = r'$\overline{x} = $' + str(result_b)
    
    for ss in x_label_b:
      soma_desv_b =+ (ss - result_b) ** 2

    dp_b = round((soma_desv_b / len(x_label_b)) ** (1/2), 6)
    
    dp_b_graf = r'$ \sigma = $' + str(dp_b)


    fig, axs = plt.subplots(1, 2, sharey=True)

    ax0 = axs[0]

    ax0.bar(x_label_a, y_label_a,  align='center', alpha=0.5, color = 'green', width=0.0005)

    ax1 = axs[1]
    ax1.bar(x_label_b, y_label_b,  align='center', alpha=0.5, color = 'blue', width=0.0005)

    axs[0].set_title('Candidato A para '+ str(n) +' amostras')

    axs[0].set_xlabel('Proporção Amostral')
    axs[0].set_ylabel('Quantidade de Repetições')
    
    axs[0].text(0.27, yg, result_a_graf, color="black")
    
    axs[0].text(0.27, yg - 5, dp_a_graf, color="black")
    
    axs[0].grid(True)


    axs[1].set_title('Candidato B para '+ str(n) +' amostras')
    axs[1].set_xlabel('Proporção Amostral')
    
    axs[1].text(0.67, yg, result_b_graf, color="black")
    
    axs[1].text(0.67, yg - 5, dp_b_graf, color="black")
    
    axs[1].grid(True)

    # axs.grid(True)

    plt.show()


ger_prob(2000, 0, 0, 70)
ger_prob(8000, 0, 0, 200)
```
